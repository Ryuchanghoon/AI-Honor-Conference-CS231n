# Reinforcement Learning

강화학습1.PNG

기본적으로 강화학습에는 행동을 취하는 Agent와 Environment가 주어진다.  
Environment가 주는 상태를 통해 행동을 취하고 그에 따른 보상을 받는 것을 반복한다.

MDP.PNG

MDP는 강화 학습 방법을 수시확하는 것을 말한다.  
* S: 가능한 상태 집합  
* A: 가능한 행동 집합  
* R: (상태, 행동)의 쌍이 주어졌을 때 보상에 대한 분포  
* P: (상태, 행동)의 쌍이 주어졌을 때 상태에 대한 분포(transition distribution)  
* $\lambda$: 보상을 받는 시간에 대한 중요도  


Value function: 어떤 상태 S와 정책 $\pi$가 주어졌을 때, 계산되는 누적 보상의 기댓값  
Q-Value function: 상태 s에서 어떤 행동을 해야 점수가 가장 좋은지에 대한 함수  
Bellman equation: 현재 상태의 value function과 next state의 value function 사이의 관계식  
Q*: 어떤 행동을 취했을 때 미래에 받을 보상의 최대치

Q-learn.PNG

Q-learning은 전이 확률과 보상을 초기에 알지 못한 상황에서  
Q-value iteration 알고리즘을 적용한 것  


FB.PNG

손실함수는 Q(s,a)가 목적함수($y_i$)랑 얼마나 차이가 나는지 측정.  
Backward Pass 과정에서는 Loss값에 대하여 $\theta$를 계속 update한다.

## Policy Gradients  
Q-Learning에도 function이 so complicated 하다는 단점이 있다.  
그래서 나온 것이 정책 경사인데,  
상태에 따른 Q-value값들을 학습시키는 것이 아니라 정책을 학습시키는 방법이다.

Policy.PNG

$J(\theta)$는 미래에 받을 보상을 누적으로 하여 기댓값을 나타낸 함수  
여기서는 Gradient Ascent 방법으로 Parameter Update를 진행한다.
