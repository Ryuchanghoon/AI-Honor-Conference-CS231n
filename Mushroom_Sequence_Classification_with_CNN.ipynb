```python
sequences = []

columns = X.columns

for idx, row in X.iterrows():
    sequence = []

    for i in range(len(columns)):
        sequence.append(row[columns[i]])
    sequences.append(sequence)

print('{sequence}: {label}'.format(sequence = sequences[0], label = y[0]))
print('len of sequences:', len(sequences[0]))
```

```python
x_train, x_test, y_train, y_test = train_test_split(sequences, y,
                                                    test_size=0.1,
                                                    random_state=RANDOM_SEED)
x_train = np.array(x_train)
x_test = np.array(x_test)
y_train = np.array(y_train)
y_test = np.array(y_test)

print(x_train)
print(y_train)
```

model에다가 add하는 방식.
<br>
GlobalMaxPooling1D는 1D 입력 데이터의 전역 최대 풀링(Global Max Pooling) 
<br>
1차원 배열 데이터에는 Conv1D를, 2차원 배열 데이터에는 Conv2D

```python
model = build_model()
model.summary()
```

```python
# Train.
history = model.fit(x_train, y_train,
                    epochs=10,
                    verbose=True,
                    validation_data=(x_test, y_test),
                    batch_size=16)
```

```python
preds_test = model.predict(x_test)
preds_test_classes = (preds_test > 0.5).astype('int32')

cnn_metrics = {'acc': metrics.accuracy_score(y_test, preds_test_classes)}
cnn_metrics['prec'] = metrics.precision_score(y_test, preds_test_classes)
cnn_metrics['rec'] = metrics.recall_score(y_test, preds_test_classes)
cnn_metrics['f1'] = metrics.f1_score(y_test, preds_test_classes)
cnn_metrics['f1_macro'] = metrics.f1_score(y_test, preds_test_classes, average='macro')
cnn_metrics['auc'] = metrics.roc_auc_score(y_test, preds_test)

for metric in cnn_metrics:
    print('{metric_name}: {metric_value}'.format(metric_name=metric, metric_value=cnn_metrics[metric]))

training_loss = history.history['loss']
test_loss = history.history['val_loss']

training_acc = history.history['accuracy']
test_acc = history.history['val_accuracy']

epoch_count = range(1, len(training_loss) + 1)

plt.plot(epoch_count, training_loss, 'r--')
plt.plot(epoch_count, test_loss, 'b-')
plt.legend(['Training Loss', 'Test Loss'])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show();

plt.plot(epoch_count, training_acc, 'r--')
plt.plot(epoch_count, test_acc, 'b-')
plt.legend(['Training Acc', 'Test Acc'])
plt.xlabel('Epoch')
plt.ylabel('Acc')
plt.show()
```
